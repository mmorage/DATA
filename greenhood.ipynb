```python
Mue=scipy.ndimage.imread('Muenchen.jpg')
Sah=scipy.ndimage.imread('Sahara.jpg')
```

```python
#calibrator:
def calib(imagen_in):
	aver_r=np.average(imagen_in[:,:,0])
	aver_g=np.average(imagen_in[:,:,1])
	aver_v=np.average(imagen_in[:,:,2])
	norm=[aver_r,aver_g,aver_v]
	return(norm)
```

```python
def greenhood_level(imagen_test,norma):
	im_tmp=(norma/imagen_test)*(-1,-1,0)
	out=-np.cos(-1*(im_tmp[:,:,0]+im_tmp[:,:,1]))
	return(out)
```

```python
implot=plt.imshow(Mue)#Muenchen
plt.axis("off")
```

```python
Mue_index=greenhood_level(Mue,norma)

implot=plt.imshow(Mue_index)#,cmap=cm.coolwarm)
plt.axis("off")
plt.colorbar(implot, orientation='horizontal')
#
plt.show()
```
